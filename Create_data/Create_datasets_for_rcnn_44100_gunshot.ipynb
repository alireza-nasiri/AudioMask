```python
import matplotlib.pyplot as plt
import csv

import librosa
import librosa.display
import numpy as np
import soundfile as sf

import scipy.misc
import os
```

```python
train_path_meta = "MyCode/data/TUT-rare-sound-events-2017-development/devtrain_15000_Prob1/649dd7c41e7bf7153d5ff159334c958d/"
names = []
names.clear()
del names[:]
with open(train_path_meta+'meta/event_list_devtrain_gunshot.csv', 'rt') as train_file:
	reader = list(csv.reader(train_file, delimiter='\n'))
	for row in reader:
		fields= str(row).split('\\t')
		tag = fields[0].replace("[\'", "")
		tag = tag.replace("\']", "")
		names.append(tag)
train_file.close()
```

```python
#creating balck and white ground truth pictures with size 128*1305
number_of_windows_per_file = 1305
def create_balckwhite_128(name):
    pic = np.zeros(shape=(128,1305), dtype=float)
    
    
    with open(train_path_meta+'/meta/event_list_devtrain_gunshot.csv', 'rt') as bc_file:
        reader = list(csv.reader(bc_file, delimiter='\n'))
        for row in reader:
            fields= str(row).split('\\t')
            name_of_file = fields[0].replace("[\'", "")
            name_of_file = name_of_file.replace("\']", "")
            if(name == name_of_file):
                if(len(fields)>=3 and fields[1]):
                    for i in range(0,number_of_windows_per_file):
                        #multiplying fileds 1 and 2 with 1000 to change them to msec
                        #I am using 23 here, because the hop_length is 23
                        if((float(fields[1])*1000)<=(i*23) and (i*23 + 46)<float(fields[2])*1000):
                            for k in range(0,128):
                                pic[k][i]=255 #because we want the balckwhite picture to be in the middle
                    
                
    bc_file.close()
    return pic
```

```python
#creating grayscale spectrograms as training data with shape (128,1292)

for i in range(0,len(names)):
    print("Creating Sepectrograms...")
    print(names[i])
    y, sr = librosa.load(train_path_meta+'audio/'+names[i], sr=44100)
    S=librosa.feature.melspectrogram(y, sr=44100, n_mels=128, n_fft=2028, hop_length=1014)
    log_S = librosa.power_to_db(S, ref=np.max)
    
    log_S =np.abs(log_S)/ np.max(np.abs(log_S),axis=0)
    log_S *= (255.0 )
    scipy.misc.imsave('train_5000_gunshot_Prob1_44100_128/'+str(i)+'.png', log_S)
    p =create_balckwhite_128(names[i])
    directory_name = 'train_5000_gunshot_Prob1_44100_128/'+str(i) 
    os.makedirs(directory_name)
    scipy.misc.imsave(directory_name+'/'+str(i)+'.png', p)
```
